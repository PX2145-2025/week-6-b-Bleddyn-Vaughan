```python
#2
#using monte carlo method to find pi 

x,y=2*np.random.rand(2,N)-1
r=np.sqrt(x*x+y*y)
n=np.sum(r <= 1.0)
print(4*n/N)
```

```python
inpond = np.sqrt(x**2+y**2) <= 1.0 # True if in the pond 
plt.figure(figsize=(6, 6))
plt.plot(x[inpond == True], y[inpond == True], 'b.', markersize=1, label='Inside circle')#plots inside circle 
plt.plot(x[inpond == False], y[inpond == False], 'g.', markersize=1, label='Outside circle')# plots if points lie outside circle 
plt.xlabel("x")
plt.ylabel("y")
plt.title("plot of points inside circle(blue) and outside circle (green)")

plt.axis('equal')
plt.show()
```

```python
#speed using for loops 

startTime = datetime.now()
N=int(1e7)
n=0
for i in range (N):
    x = 2*random()-1 
    y = 2*random()-1
    r = np.sqrt(x*x+y*y) 
    if r <= 1.0:
        n+=1 # adds 1 to n
print(4*n/N)
print(datetime.now() - startTime)
```

```python
#3
#function defined
def f(x):
    return np.sin(x)*np.tan(x)
# parameters
a = 0       # lower limit
b = np.pi/4 # upper limit
N = 1000000 # number of random points

# Generate random points betweeen a and b 
x_random = np.random.uniform(a, b, N)

# Evaluate the function at these points
y_random = f(x_random)

# Estimate the integral
integral_estimate = (b - a) * np.mean(y_random)
print(f"Monte Carlo estimate of the integral: {integral_estimate}")
```

```python
#exact answer from wolfram alpha 0.1743 so my answer and exact answer very close
```

```python
#creating random numbers 
N_values = np.arange(1000, 100001, 5000)
integral_estimates=[]# stores all values for estimate 
for N in N_values:
    x_random = np.random.uniform(a, b, N)  # vectorized random points
    y_random = f(x_random)
    estimate = (b - a) * np.mean(y_random)# creating estimates for integral
    integral_estimates.append(estimate)
#plot of graph
plt.figure(figsize=(10,6))
plt.plot(N_values,integral_estimates,color='blue')
plt.xlabel('Number of random points (N)')
plt.ylabel('Monte Carlo estimate of integral')
plt.title("number of random points agaisnt monte carlo estimate")
plt.legend()
plt.grid(True)
plt.show()
    
```
